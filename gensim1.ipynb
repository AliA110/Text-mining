## split sentences to word

### tokenize

```python
texts=[[text for text in doc.split()]for doc in documents]
```

```python
dictionary=corpora.Dictionary(texts)
```

```python
print (dictionary)
```

### show word id in map

```python
display(dictionary.token2id)
```

#### add text 2 to the dictionery

```python
texts2=[[text for text in doc.split()]for doc in documents_2]
dictionary.add_documents(texts2)
```

### print dictionery

```python
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
texts=texts+texts2
```

```python
mydic=corpora.Dictionary()
mycorpus=[mydic.doc2bow(doc, allow_update=True) for doc in texts]
pprint(mycorpus)
```

### showing word whit number of words

```python
word_counts=[[(mydic[id],count)for id,count in line ]for line in mycorpus]
pprint(word_counts)
```

## Use tf idf to weight words

```python
tfidf=models.TfidfModel(mycorpus,smartirs='ntc')
```

### show the word wieghs in corpus

```python
for doc in mycorpus:
    print([[dictionary[id],freq]for id ,freq in doc])
```

### show tf_idf weights

```python
for sent in tfidf[mycorpus]:
    print([[dictionary[id],np.around(frequency,decimals=2)]for id,frequency in sent])
```

### creat vector

```python
model=Word2Vec(mycorpus,min_count=0,workers=cpu_count())
```
